## 1. csvファイルの読み込み

```python
import pandas as pd
import glob

# 同じフォルダー内に入れておいた9つのcsvファイルの名前を取得
allFiles = glob.glob('*.csv') 
allFiles
```

## 3. 全csvを連結してデータセットを保存

```python
# 日付を基準に結合
merged = pd.merge(df0,df1,on='日付', how='outer')
merged = pd.merge(merged,df2,on='日付', how='outer')
merged = pd.merge(merged,df3,on='日付', how='outer')
merged = pd.merge(merged,df5,on='日付', how='outer')
merged = pd.merge(merged,df6,on='日付', how='outer')
merged = pd.merge(merged,df7,on='日付', how='outer')
merged = pd.merge(merged,df8,on='日付', how='outer')
merged = pd.merge(merged,df4,on='日付', how='outer')
```

```python
# データセットに必要な期間は2020/02/04 ~ 2020/11/28
dataset = merged[:-82]
dataset
```

```python
# automlでの予測用データとしてcsvに保存
dataset.to_csv('automl_data.csv', index=False)
```

```python
# python(sarima)での予測用データは'日付', 'PCR 検査陽性者数(単日)'の2列だけ
sarima = dataset[['日付', 'PCR 検査陽性者数(単日)']]
sarima
```
